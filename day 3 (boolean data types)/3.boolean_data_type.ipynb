<h2><code style="background:Gold;color:black">2)Conditional Statements: :-</code></h2>
<h4><pre>
            Boolean values are extensively used in conditional statements to 
            control the flow of your program. 

 </pre></h4>           
<h4><pre>
                Types of Control Statement:

                        1) if: 
                                Executes a block of code if a condition is True.
                                
                        2) elif (short for "else if"):
                                 Executes a block of code if another condition is 
                                 True (and the previous conditions were not met).
                                 
                        3) else: Executes a block of code if none of the previous
                         conditions were met (i.e., when the condition is False).
</pre></h4>

```python
x = 5
y = 10


if (x<y):
    print("x is less than y")
elif (x>y):
    print("x is greater than y")
else:
    print("")
```

<h2><code style="background:Gold;color:black">3) Logical Operations :-</code></h2>
<h4><pre>
            Logical operations involve combining and manipulating boolean 
            values to create more complex conditions. 

<pre>

                     Python provides three main logical operators:
                     
                           1) and: 
                                    Returns True if both operands are True.
                           2) or: 
                                    Returns True if at least one of the 
                                    
                                    operands is True.
                           3) not: 
                                    Returns the opposite of the operand's 
                                    boolean value.
</pre></h4>

```python
x = 5
y = 10
z = 7

# Using 'and' operator
if x > 0 and y > 0:
    print("Both x and y are greater than 0")

# Using 'or' operator
if x < 0 or y < 0:
    print("At least one of x and y is less than 0")

# Using 'not' operator
if not (z == 7):
    print("z is not equal to 7")
```

<pre>            1) is and is not: 
                These operators are used to compare object identity, 
                which means they check if two objects reference the same memory location. 
                They are typically used when comparing objects such as variables or instances.
</pre>

```python
x = 5
y = x  # y refers to the same integer object as x

print(x is y)    # True, x and y refer to the same integer object
print(x is not y)  # False, x and y have the same identity
```

<pre>               2) in and not in: These operators are used to check if a value is present 
                    in a sequence, such as a list, tuple, string, or dictionary. 
                    They are commonly used in conditional statements and loops.

```python

name=["vikas","kumar","rohit","syam"]
n="devesh" in name
print(n)

```

<pre>                3) all() and any(): 
                        These functions are used to perform logical operations on an 
                        iterable (e.g., a list) and return a boolean result. 
                        
                        1)all() returns True if all elements in the iterable are True 
                        
                        2)any() returns True if at least one element in the iterable is True

```python
my_list = [True, False, True, True]

result_all = all(my_list)  # False, because not all elements are True
result_any = any(my_list)  # True, because at least one element is True
print(result_all)
print(result_any)
```

```python
my_list = [True, True, True, True]

r=any(my_list)
print(r)
rr=all(my_list)
print(rr)
```
