<code style="background:Gold;color:black">Escape Characters::-</code>
<pre>
    Escape characters in Python are special characters that are used to 
    represent characters that are impossible to include directly in a string

    They are preceded by a backslash (\) and indicate to Python that the character 
    following the backslash should be treated differently.

        # example of Escaping character
            1)backslash
            2)single quote
            3)souble quote
            4)newline


</pre>

```python
# string using single quotes

a='xyz'
print(a)
```

```python
print(type(a))
```

```python
# string using double quotes
b="xyz"
print(b)
```

```python
print(type(b))
```

```python
# string using sinle triplle quotes
c='''xyz'''
print(c)
```
